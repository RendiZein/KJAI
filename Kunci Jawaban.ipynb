# 2 (Candy)

```python
# Data jumlah permen
candy_names = ['Kit Kat', 'Snickers', 'Milky Way', 'Toblerone', 'Twix']
candy_counts = [52, 39, 78, 13, 78]
explode = (0, 0.1, 0, 0, 0)

total_candies = sum(candy_counts)
candy_prob = [count/total_candies for count in candy_counts]

# Chart
plt.figure(figsize=(10,6))
plt.pie(candy_prob, labels=candy_names, autopct='%1.1f%%', explode=explode)
plt.title('Peluang Memilih Setiap Jenis Permen')
plt.show()
```

# 3 (Makanan)

```python
# Data penjualan makanan penutup
dessert_sales = {
    'Lava Cake': 14,
    'Mousse': 5,
    'Chocolate Cake': 12,
    'Ice Cream': 19,
    'Truffles': 6,
    'Brownie': 8,
    'Chocolate Chip Cookie': 12,
    'Chocolate Pudding': 9,
    'Souffle': 10,
    'Chocolate Cheesecake': 17,
    'Chocolate Chips': 2,
    'Fudge': 9,
    'Mochi': 13,
}


sorted_dessert_sales = sorted(dessert_sales.items(), key=lambda item: item[1])

#Selection Method
colors = ['red' if i < 3 else 'green' for i in range(len(sorted_dessert_sales))]

plt.figure(figsize=(10,6))
plt.barh([item[0] for item in sorted_dessert_sales], [item[1] for item in sorted_dessert_sales], color=colors)
plt.title('Penjualan Makanan Penutup')
plt.xlabel('Penjualan')
plt.ylabel('Makanan Penutup')
plt.show()
```
